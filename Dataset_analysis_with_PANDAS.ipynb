```python
from data_utils import metadata_loader
from config import Config
import pandas as pd
import matplotlib.pyplot as plt
```

```python
config = Config()
metadata = metadata_loader.MetadataLoader(config).load_metadata()
```

```python
md_df = pd.DataFrame(metadata['valid']).drop(['object_labels', 'set_name']).T.astype(int)
print(md_df.head())

md_df.describe()
```

```python
plt.figure(figsize=(20,9))
md_df.action_label.hist(bins=175)
_=plt.title("Frequency of classes in the dataset", fontweight="bold", size = 35)
_=plt.xlabel('class',fontweight="bold", size = 20)
_=plt.ylabel('total times',fontweight="bold", size = 20)
plt.tick_params(labelsize=20)
```

```python
plt.figure(figsize=(20,9))
hist = md_df.n_frames.hist(bins = 20)
_=plt.title("Frequency of video length", fontweight="bold", size = 35)
_=plt.xlabel('nr of frames',fontweight="bold", size = 20)
_=plt.ylabel('total times',fontweight="bold", size = 20)
plt.tick_params(labelsize=20)
```

```python
plt.figure(figsize=(20,9))
hist = md_df.n_frames.hist(bins = 20, cumulative=True)
_=plt.title("Frequency of video length", fontweight="bold", size = 35)
_=plt.xlabel('nr of frames',fontweight="bold", size = 20)
_=plt.ylabel('total times',fontweight="bold", size = 20)
plt.tick_params(labelsize=20)
```
